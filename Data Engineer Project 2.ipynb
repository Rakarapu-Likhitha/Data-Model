```python
import mysql.connector
import pandas as pd
```

```python
def create_database():
      # Connect to default database
      conn = mysql.connector.connect(
          host="127.0.0.1",
          user="root",
          password="Andhra@1234",
          use_pure=False  # Enable cext option
      )
      cur = conn.cursor()
    
    # Create sparkify database with UTF8 encoding
      cur.execute("DROP DATABASE IF EXISTS accounts")
      cur.execute("CREATE DATABASE accounts CHARACTER SET utf8mb4 COLLATE utf8mb4_unicode_ci")
    
      # Switch to the newly created database
      cur.execute("USE accounts")
    
      return cur, conn
```
